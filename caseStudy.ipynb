```python
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df['Release']= pd.to_datetime(df['Release'])
df.dtypes
```

```python
# handdle missing value
# df['Series']=df['Series'].fillna(df['Series'].mode()[0])
# print(df.isnull().sum())

df['Series']=df['Series'].mask(df['Series'].isna(), df['Name'])
print(df.isnull().sum())
```

```python
# check duplicated rows
df_duplicated= df[df.duplicated()]
df_duplicated
```

```python
# drop duplicated
df = df.drop_duplicates(keep=False)
```

```python
df_cleaned = df.copy()
```

# Which game is the oldest and the newest games in that dataset?

```python
#cara 1
df_cleaned.sort_values(by=['Release', 'Name'], inplace=True)
df_cleaned
```

```python
#cara 2 = the oldest games
df_cleaned[['Name', 'Release']].sort_values('Release').head()
```

```python
#cara 2 = the newest games
df_cleaned[['Name', 'Release']].sort_values('Release', ascending=False).head()
```

# Which publisher published most of the games?



```python
df_publisher = df_cleaned.groupby(['Publisher'])['Publisher'] \
                      .count()  \
                      .reset_index(name='count') \
                      .sort_values(['count'], ascending=False)

df_publisher
```

```python
#cara lain
df_publisher = df_cleaned.groupby('Publisher', as_index=False)['Name'].nunique()
df_publisher.sort_values('Name', ascending=False)
```

```python
plt.rcParams["figure.figsize"] = (18,8)
sns.barplot('Publisher', 'Name', data = df_publisher.sort_values('Name', ascending=False).head(10))
```

publisher yang menerbitkan sebagian besar game adalah Electronic Arts

## Which developer developed most of the games?

```python
df_dev = df_cleaned.groupby(['Developer'])['Developer'] \
                      .count()  \
                      .reset_index(name='count') \
                      .sort_values(['count'], ascending=False)
df_dev
```

Blizzard Entertainment menjadi developer yang paling banyak mengembangakan game
